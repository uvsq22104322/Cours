## Module matplotlib

Le module **matplotlib** permet de générer des graphiques depuis Python. Il est l’outil complémentaire de ***NumPy***, ***scipy*** ou ***pandas***.

### Représentation sous forme de points

Considérons comme exemple *l’évolution de la concentration d’un produit dans le sang (exprimé en mg/L) en fonction du temps (exprimé en heure)*. Nous avons les résultats suivants :

| ![concentration d’un produit dans le sang](images/cpsang.png) |
|----|

> Nous allons maintenant représenter l’évolution de la concentration en fonction du temps :


````python
# On importe le sous-module pyplot du module matplotlib et on lui donne l’alias plt 
# pour l’utiliser plus rapidement ensuite 
# (cet alias est standard, utilisez-la systématiquement).
import matplotlib.pyplot as plt

# On définit les variables temps et concentration comme des listes. 
# Les deux listes doivent avoir la même longueur (7 éléments dans le cas présent).
temps=[1,2,3,4,6,7,9]
concentration = [5.5, 7.2, 11.8, 13.6, 19.1, 21.7, 29.4]

# La fonction scatter() permet de représenter des points sous forme de nuage de points. 
# Les deux premiers arguments correspondent aux valeurs en abscisse et en ordonnée des points, 
# fournis sous forme de listes. 
# Des arguments facultatifs sont ensuite précisés comme le symbole (marker) et la couleur (color).
plt.scatter(temps, concentration, marker="o", color="blue")

# Les fonctions xlabel() et ylabel() sont utilisées pour donner un nom aux axes.
plt.xlabel("Temps (h)")
plt.ylabel("Concentration (mg/L)")

# La fonction title() définit le titre du graphique.
plt.title("Concentration de produit en fonction du temps")

# la fonction show() affiche le graphique généré à l’écran.
plt.show()
````

* Tout d’abord, on importe le sous-module pyplot du module matplotlib et on lui donne l’alias plt pour l’utiliser plus rapidement ensuite (cet alias est standard, utilisez-la systématiquement).
* On définit les variables temps et concentration comme des listes. Les deux listes doivent avoir la même longueur (7 éléments dans le cas présent).
* La fonction scatter() permet de représenter des points sous forme de nuage de points. Les deux premiers arguments correspondent aux valeurs en abscisse et en ordonnée des points, fournis sous forme de listes. Des arguments facultatifs sont ensuite précisés comme le symbole (marker) et la couleur (color).
* Les fonctions xlabel() et ylabel() sont utilisées pour donner un nom aux axes. 
* La fonction title() définit le titre du graphique.
* Enfin, la fonction show() affiche le graphique généré à l’écran.


### Représentation sous forme de courbe

On sait par ailleurs que l’évolution de la concentration du produit en fonction du temps peut-être modélisée par la fonction ```f (x) = 2 + 3x```. Représentons ce modèle avec les points expérimentaux et sauvegardons le graphique obtenu sous forme d’une image :

````python
# On charge le module numpy sous le nom np.
import numpy as np

# On importe le sous-module pyplot du module matplotlib et on lui donne l’alias plt 
import matplotlib.pyplot as plt

# On définit les variables temps et concentration comme des listes. 
temps=[1,2,3,4,6,7,9]
concentration = [5.5, 7.2, 11.8, 13.6, 19.1, 21.7, 29.4]

# On utilise la fonction scatter() pour représenter des points sous forme de nuage de points. 
plt.scatter(temps, concentration, marker="o", color = "blue")

# On utilise les fonctions xlabel() et ylabel() sont utilisées pour donner un nom aux axes.
plt.xlabel("Temps (h)")
plt.ylabel("Concentration (mg/L)")

# La fonction title() pour définir le titre du graphique.
plt.title("Concentration de produit en fonction du temps")

# On crée la variable x avec la fonction linspace() du module NumPy qui renvoie une liste 
# de valeurs régulièrement espacées entre deux bornes, 
# ici entre le minimum (min(temps)) et le maximum (max(temps)) de la variable temps. 
# Dans notre exemple, nous générons une liste de 50 valeurs. 
# La variable x ainsi créée est du type array.
x = np.linspace(min(temps), max(temps), 50)

# On construit ensuite la variable y à partir de la formule modélisant l’évolution 
# de la concentration en fonction du temps. 
# Cette manipulation n’est possible que parce que x est du type array. 
# Cela ne fonctionnerait pas avec une liste classique. 
y = 2 + 3 * x

# La fonction plot() permet de construire une courbe à partir des coordonnées 
# en abscisse et en ordonnées des points à représenter. 
# On indique ensuite des arguments facultatifs comme le style de la ligne (ls) et sa couleur (color). 
plt.plot(x, y, color='green', ls="--")

# La fonction grid() affiche une grille.
plt.grid()

# Enfin, la fonction savefig() enregistre le graphique produit sous la forme d’une image au format png. 
# Des arguments par mot-clé définissent la manière de générer les marges autour du graphique (bbox_inches) 
# et la résolution de l’image (dpi).
plt.savefig("concentration_vs_temps.png", bbox_inches='tight', dpi=200)

````

Les étapes supplémentaires par rapport au graphique précédent sont :
* On charge le module numpy sous le nom np.
* On crée la variable x avec la fonction linspace() du module NumPy qui renvoie une liste de valeurs régulièrement espacées entre deux bornes, ici entre le minimum (min(temps)) et le maximum (max(temps)) de la variable temps. Dans notre exemple, nous générons une liste de 50 valeurs. La variable x ainsi créée est du type array.
* On construit ensuite la variable y à partir de la formule modélisant l’évolution de la concentration en fonction du temps. Cette manipulation n’est possible que parce que x est du type array. Cela ne fonctionnerait pas avec une liste classique. 
* La fonction plot() permet de construire une courbe à partir des coordonnées en abscisse et en ordonnées des points à représenter. On indique ensuite des arguments facultatifs comme le style de la ligne (ls) et sa couleur (color). 
* La fonction grid() affiche une grille.
* Enfin, la fonction savefig() enregistre le graphique produit sous la forme d’une image au format png. Des arguments par mot-clé définissent la manière de générer les marges autour du graphique (bbox_inches) et la résolution de l’image (dpi).


### Représentation sous forme de diagramme en bâtons

On souhaite maintenant représenter graphiquement la distribution des différentes bases dans une séquence d’ADN.

````python
import numpy as np
import matplotlib.pyplot as plt

# On définit les variables sequence, bases et distribution qu’on utilise ensuite.
sequence = "ACGATCATAGCGAGCTACGTAGAA"
bases = ["A", "C", "G", "T"]
distribution = []

# On calcule la distribution des différentes bases dans la séquence. 
# On utilise pour cela la méthode count() qui renvoie le nombre de fois 
# qu’une chaîne de caractères (les différentes bases) se trouve dans une autre (la séquence). 
for base in bases:
    distribution.append(sequence.count(base))

# On définit la position en abscisse des barres. Dans cet exemple, la variable x vaut array([0, 1, 2, 3])
x = np.arange(len(bases))

# La fonction bar() construit le diagramme en bâtons. 
# Elle prend en argument la position des barres (x) et 
# leurs hauteurs (distribution).
plt.bar(x, distribution)

# La fonction xtics() redéfinit les étiquettes (c’est-à-dire le nom des bases) sur l’axe des abscisses.
plt.xticks(x, bases)

# On définit les légendes des axes et le titre du graphique. 
# On insère un retour à la ligne \n dans le titre pour qu’il 
# soit réparti sur deux lignes.
plt.xlabel("Bases")
plt.ylabel("Nombre")
plt.title(f"Distribution des bases\n dans la séquence {sequence}")

# Enfin, on enregistre le graphique généré au format png.
plt.savefig("distribution_bases.png", bbox_inches="tight", dpi=200)
````

Examinant les différentes étapes du script précédent :
* On définit les variables sequence, bases et distribution qu’on utilise ensuite.
* On calcule la distribution des différentes bases dans la séquence. On utilise pour cela la méthode count() qui renvoie le nombre de fois qu’une chaîne de caractères (les différentes bases) se trouve dans une autre (la séquence). 
* On définit la position en abscisse des barres. Dans cet exemple, la variable x vaut ```array([0, 1, 2, 3])```. 
* La fonction bar() construit le diagramme en bâtons. Elle prend en argument la position des barres (x) et leurs hauteurs (distribution).
* La fonction xtics() redéfinit les étiquettes (c’est-à-dire le nom des bases) sur l’axe des abscisses.
* On définit les légendes des axes et le titre du graphique. On insère un retour à la ligne \n dans le titre pour qu’il soit réparti sur deux lignes.
* Enfin, on enregistre le graphique généré au format png.
