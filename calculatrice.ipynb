```python
import tkinter as tk
import math as meth

def Un():
    txt = label.cget("text")
    label.config(text=txt + "1")

def Deux():
    txt = label.cget("text")
    label.config(text=txt + "2")

def Trois():
    txt = label.cget("text")
    label.config(text=txt + "3")

def Quatre():
    txt = label.cget("text")
    label.config(text=txt + "4")

def Cinq():
    txt = label.cget("text")
    label.config(text=txt + "5")

def Six():
    txt = label.cget("text")
    label.config(text=txt + "6")

def Sept():
    txt = label.cget("text")
    label.config(text=txt + "7")

def Huit():
    txt = label.cget("text")
    label.config(text=txt + "8")

def Neuf():
    txt = label.cget("text")
    label.config(text=txt + "9")

def Zéro():
    txt = label.cget("text")
    label.config(text=txt + "0")

def Plus():
    txt = label.cget("text")
    label.config(text=txt + "+")

def Moins():
    txt = label.cget("text")
    label.config(text=txt + "-")

def Fois():
    txt = label.cget("text")
    label.config(text=txt + "*")

def Diviser():
    txt = label.cget("text")
    label.config(text=txt + r"/")

def Calculer():
    x = label.cget("text")
    result = eval(x)
    label.config(text=str(result))

def reset():
    label.config(text=" ")


root = tk.Tk()

root.grid_columnconfigure( weight=1)
root.grid_rowconfigure( weight=1)

label = tk.Label(root, text=" ")

bCalculer = tk.Button(root, text="=", command=Calculer)
bUn = tk.Button(root, text="1", command=Un)
bDeux = tk.Button(root, text="2", command=Deux)
bTrois = tk.Button(root, text="3", command=Trois)
bQuatre = tk.Button(root, text="4", command=Quatre)
bCinq = tk.Button(root, text="5", command=Cinq)
bSix = tk.Button(root, text="6", command=Six)
bSept = tk.Button(root, text="7", command=Sept)
bHuit = tk.Button(root, text="8", command=Huit)
bNeuf = tk.Button(root, text="9", command=Neuf)
bZéro = tk.Button(root, text="0", command=Zéro)
bPlus = tk.Button(root, text="+", command=Plus)
bMoins = tk.Button(root, text="-", command=Moins)
bFois = tk.Button(root, text="*", command=Fois)
bDiviser = tk.Button(root, text=r"/", command=Diviser)
bReset = tk.Button(root, text="Clear", command=reset)


label.grid(row=0,column=0, rowspan=2,columnspan=4)
bCalculer.grid(row=5,column=3)
bUn.grid(row=4,column=0)
bDeux.grid(row=4,column=1)
bTrois.grid(row=4,column=2)
bQuatre.grid(row=3,column=0)
bCinq.grid(row=3,column=1)
bSix.grid(row=3,column=2)
bSept.grid(row=2,column=0)
bHuit.grid(row=2,column=1)
bNeuf.grid(row=2,column=2)
bZéro.grid(row=5,column=1)
bPlus.grid(row=4,column=3)
bMoins.grid(row=3,column=3)
bFois.grid(row=2,column=3)
bDiviser.grid(row=5,column=2)
bReset.grid(row=5, column=0)

root.mainloop()
```
